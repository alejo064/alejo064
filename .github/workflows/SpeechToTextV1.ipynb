```python
from ibm_watson import SpeechToTextV1
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
```

```python
apikey = 'eR5oSSWcjra0z6F82ilbg4KbxqbX1owcvDARNjxRlC1o'
url = 'https://api.us-south.speech-to-text.watson.cloud.ibm.com/instances/a7607c47-dbf6-452a-a0d5-82e3ea910836'
```

```python
# Setup Service
authenticator = IAMAuthenticator(apikey)
stt = SpeechToTextV1(authenticator=authenticator)
stt.set_service_url(url)
```

```python
# perform conversion
with open('d:/Untitled.mp3', 'rb') as f:
    res = stt.recognize(audio=f, content_type='audio/mp3', model='es-CO_BroadbandModel', continous=True).get_result()
```

```python
res
```

```python
text = res['results'][0]['alternatives'][0]['transcript']
text
```
